#Estadistica Descriptiva

```python
import pandas as pd

# Lectura del archivo CSV
df = pd.read_csv("/content/drive/MyDrive/Escuela/mineria de datos/datos_limpio.csv")

# Creación del DataFrame con las columnas seleccionadas
df_LIMP = pd.DataFrame.from_dict(df.loc[:,['ID','Name','Average User Rating','User Rating Count','Price',
                                           'In-app Purchases','Developer','Age Rating','Primary Genre',
                                           'Genres','Original Release Date','Current Version Release Date']])

print(df_LIMP)
```

```python
# Cálculo del promedio de la calificación de usuarios
mean_rating = df_LIMP['Average User Rating'].mean()
print(f"Promedio de la calificación de usuarios: {mean_rating}")


```

```python
# Cálculo del promedio de la cantidad de usuarios que han calificado
mean_user_count = df_LIMP['User Rating Count'].mean()
print(f"Promedio de la cantidad de usuarios que han calificado: {mean_user_count}")

```

```python
# Cálculo de la moda del precio
mode_price = df_LIMP['Price'].mode()
print(f"Moda del precio: {mode_price}")

```

```python
# Cálculo del precio máximo y mínimo
max_price = df_LIMP['Price'].max()
min_price = df_LIMP['Price'].min()
print(f"Precio máximo: {max_price}, Precio mínimo: {min_price}")

```

```python
# Estadísticas de fecha de lanzamiento
original_release_stats = df_LIMP['Original Release Date'].describe()
print(f"Estadísticas de fecha de lanzamiento: \n{original_release_stats}")

```

```python
# Estadísticas de fecha de última versión
last_version_release_stats = df_LIMP['Current Version Release Date'].describe()
print(f"Estadísticas de fecha de última versión: \n{last_version_release_stats}")
```

#Visualizacion de datos

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from google.colab import drive

# Monta Google Drive
drive.mount('/content/drive')

# Carga los datos desde Google Drive
df = pd.read_csv("/content/drive/MyDrive/Escuela/mineria de datos/datos_limpio.csv")

# Directorio donde se guardarán las imágenes
directory = '/content/drive/MyDrive/Escuela/mineria de datos/img'

# Si el directorio no existe, se crea
if not os.path.exists(directory):
   os.makedirs(directory)

# Gráfico de barras para las primeras 15 aplicaciones y sus puntajes promedio de usuarios
df_15 = df.head(15)
fig, ax = plt.subplots()

x = df_15['Name']
y = df_15['Average User Rating']
ax.bar(x, y)

ax.set_xlabel("Aplicaciones")
ax.set_ylabel("Puntaje Promedio de Usuarios")
ax.set_title("Puntajes Promedio de las Primeras 15 Aplicaciones")

plt.xticks(rotation=90)

# Guardar la imagen en la ruta especificada
img_path = os.path.join(directory, 'puntajes_promedio.png')
plt.savefig(img_path)
plt.show()

# Gráfico de barras para las primeras 15 aplicaciones y la cantidad de calificaciones de usuarios
fig, ax = plt.subplots()

x = df_15['Name']
y = df_15['User Rating Count']
ax.bar(x, y)

ax.set_xlabel("Aplicaciones")
ax.set_ylabel("Cantidad de Calificaciones de Usuarios")
ax.set_title("Calificaciones de Usuarios de las Primeras 15 Aplicaciones")

plt.xticks(rotation=90)

# Guardar la imagen en la ruta especificada
img_path = os.path.join(directory, 'calificaciones_usuarios.png')
plt.savefig(img_path)
plt.show()

```

```python
import pandas as pd
import plotly.graph_objects as go

# Cargar datos desde Google Drive
df = pd.read_csv("/content/drive/MyDrive/Escuela/mineria de datos/datos_limpio.csv")

# Seleccionar columnas relevantes
relevant_columns = [
    'ID', 'Name', 'Average User Rating', 'User Rating Count',
    'Price', 'In-app Purchases', 'Developer', 'Age Rating',
    'Primary Genre', 'Genres', 'Original Release Date',
    'Current Version Release Date'
]

# Crear DataFrame con las columnas seleccionadas
df_LIMP = df[relevant_columns]

# Imprimir los datos
print(df_LIMP)

# Seleccionar las primeras 15 filas
df_LIMP_15 = df_LIMP.head(15)
print(df_LIMP_15)

# Crear el gráfico con Plotly
fig = go.Figure()

# Box Plot para 'Average User Rating'
fig.add_trace(go.Box(y=df_LIMP_15['Average User Rating'], name='Average User Rating'))

# Box Plot para 'User Rating Count'
fig.add_trace(go.Box(y=df_LIMP_15['User Rating Count'], name='User Rating Count'))

# Configurar el diseño del gráfico
fig.update_layout(
    title={
        'text': "Box Plots - User Ratings",
        'y': 0.9,
        'x': 0.5,
        'xanchor': 'center',
        'yanchor': 'top'
    }
)

# Mostrar el gráfico
fig.show()

# Guardar el gráfico en la ubicación especificada
fig.write_image('/content/drive/MyDrive/Escuela/mineria de datos/img/boxplots_plotly.png')





```
